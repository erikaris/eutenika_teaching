Use above credentials to authenticate ```tweepy.API```. Before to do it, we have to make sure to install tweepy package ```pip install tweepy```

````python
import tweepy

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)
````

Make a search trough that API. We provide ***-filter:retweets*** in the search term to filter out the retweet tweet. Option ```tweet_mode='extended'``` is used to get the full text of tweet instead of 160 chars truncated tweet. Since twitter apply rate limit, we should provide ```wait_on_rate_limit=True``` option.

````python
query = 'sensus penduduk 2020 -filter:retweets'
max_tweets = 1000

searched_tweets = [status.full_text.replace('\n', ' ') for status in
                   tweepy.Cursor(api.search, q=query, tweet_mode='extended',
                   wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
                   .items(max_tweets)]
````

Dump to text

````python
with open('sensus_penduduk_tweets.txt', 'w+') as f:
    f.write('\n'.join(searched_tweets))
````
